# 1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

#### 1. Artificial Intelligence (AI)

- Definition: The broad field of creating machines or systems that can perform tasks that normally require human intelligence.

- Scope: Includes reasoning, problem-solving, understanding natural language, perception, and decision-making.

- Examples: Chatbots, recommendation systems, self-driving cars, voice assistants (Siri, Alexa).


#### 2. Machine Learning (ML)

- Definition: A subset of AI that focuses on building systems that can learn from data and improve their performance over time without being explicitly programmed.

- Scope: Works with algorithms that find patterns in data and make predictions.

- Examples: Spam email detection, product recommendation, fraud detection.

#### 3. Deep Learning (DL)

- Definition: A subset of ML that uses artificial neural networks with many layers to model complex patterns in data.

- Scope: Excellent for handling large, unstructured data (images, audio, video, text).

- Examples: Face recognition, voice recognition, self-driving car vision systems, generative AI (like me!).

#### 4. Data Science (DS)

- Definition: An interdisciplinary field that focuses on extracting insights and knowledge from data using statistics, programming, and domain expertise.

- Scope: Includes data collection, cleaning, analysis, visualization, and building models (AI/ML may be part of it).

- Examples: Analyzing sales trends, predicting customer behavior, healthcare analytics.


# 2. What are the types of machine learning? Describe each with one real-world example.

#### 1. Supervised Learning

- Definition: The model is trained on labeled data (input + correct output are given).

- Goal: Learn a mapping between input and output so it can predict outputs for new, unseen inputs.

- Techniques: Regression (predicting continuous values), Classification (predicting categories).

- Example:
    Email Spam Detection → The model is trained with emails labeled as spam or not spam. Later, it predicts whether a new email is spam.

#### 2. Unsupervised Learning

- Definition: The model is trained on unlabeled data (only input data, no output labels).

- Goal: Find hidden patterns, structures, or relationships in the data.

- Techniques: Clustering, Dimensionality Reduction.

- Example:
    Customer Segmentation in Marketing → Grouping customers into clusters (e.g., frequent buyers, seasonal buyers) without prior labels, so businesses can target each group differently.

#### 3. Reinforcement Learning (RL)

- Definition: The model (called an agent) learns by interacting with an environment. It takes actions, receives rewards/penalties, and improves strategy over time.

- Goal: Maximize cumulative rewards by trial and error.

- Techniques: Q-Learning, Deep Reinforcement Learning.

- Example:
        Self-Driving Cars → The car (agent) learns to drive by receiving rewards for staying on the road and penalties for accidents.


# 3. Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

#### 1. Overfitting

- Definition: When a model learns too much from training data, including noise and random fluctuations, making it perform poorly on new/unseen data.

- Cause: Model is too complex (too many features, deep trees, too many layers, etc.).

- Symptoms:

   - Very high accuracy on training data

   - Very low accuracy on test/validation data

- Example: A decision tree that memorizes every detail of training data but fails to generalize to new samples.

#### 2. Underfitting

- Definition: When a model is too simple and cannot capture the underlying patterns in the data.

- Cause: Model has high bias (e.g., using linear regression for highly non-linear data).

- Symptoms:

     - Poor performance on both training and test data

- Example: Using a straight line to fit data that clearly follows a curve.

#### 3. Bias-Variance Tradeoff

- Bias: Error due to overly simplistic assumptions in the model (leads to underfitting).

- Variance: Error due to too much sensitivity to training data (leads to overfitting).

- Tradeoff:

    - Increasing model complexity → lowers bias but increases variance.

    - Decreasing model complexity → lowers variance but increases bias.

- Goal: Find the optimal balance where both bias and variance are minimized, ensuring good generalization.


# 4. What are outliers in a dataset, and list three common techniques for handling them.

#### Outliers in a Dataset

- Definition: Outliers are data points that deviate significantly from the majority of the data.

- Cause: Could be due to measurement errors, data entry mistakes, or genuine rare events.

- Effect: Outliers can skew results, distort statistical analyses, and reduce model performance.

#### Techniques for Handling Outliers:

1. Removal (Deleting Outliers)

- Simply remove extreme values if they are due to errors or irrelevant to the analysis.


2. Transformation

- Apply mathematical transformations (e.g., log, square root, Box-Cox) to reduce the effect of outliers.


3. Capping (Winsorization)

- Replace extreme outliers with a maximum/minimum threshold (e.g., 1st and 99th percentile values).


# 5. Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.
   
#### Handling Missing Values in a Dataset

1. Identify Missing Values

Check for missing entries using tools like isnull() or info() in pandas.

2. Analyze the Pattern

- MCAR (Missing Completely at Random): No specific reason, safe to impute.

- MAR (Missing at Random): Related to other variables.

- MNAR (Missing Not at Random): Missingness depends on the value itself (harder to handle).

3. Handle the Missing Values

- Deletion: Drop rows/columns with too many missing values.

- Imputation: Fill missing values with estimates.

- Model-based methods: Predict missing values using machine learning models.

* Imputation Techniques :
  
    - For Numerical Data → Mean/Median Imputation

        - Replace missing values with the mean or median of the column.

        - Example: If a column "Age" has missing values, replace them with the average (mean) or median age.

    - For Categorical Data → Mode Imputation

        - Replace missing values with the most frequent category (mode).

        - Example: If "Gender" has missing entries, fill them with the most common value ("Female" or "Male"

```python
#8. Write a Python script to:
   
#● Generate 1000 samples from a normal distribution.

#● Introduce 50 random missing values.

#● Fill missing values with the column mean.

#● Plot a histogram before and after imputation.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(1000, 50, replace=False)
data_with_missing = data.copy()
data_with_missing[missing_indices] = np.nan

# Convert to DataFrame for easier handling
df = pd.DataFrame({'Values': data_with_missing})

# Step 3: Fill missing values with the column mean
mean_value = df['Values'].mean()
df['Imputed'] = df['Values'].fillna(mean_value)

# Step 4: Plot histograms before and after imputation
plt.figure(figsize=(10, 5))

# Before imputation
plt.subplot(1, 2, 1)
plt.hist(df['Values'].dropna(), bins=30, edgecolor='black')
plt.title("Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1, 2, 2)
plt.hist(df['Imputed'], bins=30, edgecolor='black')
plt.title("After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
